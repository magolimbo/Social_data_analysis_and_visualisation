```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('Police_Department_Incident_Reports__Historical_2003_to_May_2018_20240129.csv')
data.head()
```

```python
print("total number of crimes: ", len(data))


category_counts = data['Category'].value_counts()
print("total number of types of crimes: ", len(category_counts))
print(category_counts)


most_common_category = category_counts.idxmax()
print("Crime category with the most instances:", most_common_category)

least_common_category = category_counts.idxmin()
print("Crime category with the least instances:", least_common_category)



```

```python
categories = category_counts.index
instances = category_counts.values
plt.bar(categories, instances)
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract the year from the "Date" column
data['Year'] = pd.to_datetime(data['Date']).dt.year

# Count the occurrences for each year
year_counts = data['Year'].value_counts()

# Find the year with the most crimes
most_common_year = year_counts.idxmax()
print("Year with the most crimes:", most_common_year)

# Find the year with the fewest crimes
least_common_year = year_counts.idxmin()
print("Year with the fewest crimes:", least_common_year)


hist_years = plt.bar(year_counts.index, year_counts.values)
plt.xlabel('Year')
plt.ylabel('Number of crimes')
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
# Set the font family and size
mpl.rcParams['font.family'] = 'helvetica'
mpl.rcParams['font.size'] = 12

# Imposta lo stile "darkgrid" di seaborn
sns.set(style="darkgrid")

focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])

for crime in focuscrimes:
    spec_crime_row = data[data['Category'] == crime]
    occurrences_crime = spec_crime_row['Year'].value_counts()
    
    # Personalizza il plot
    plt.figure(figsize=(10, 6))  # Imposta la dimensione della figura
    sns.barplot(x=occurrences_crime.index, y=occurrences_crime.values, palette="cubehelix")  # Utilizza la palette "husl"
    plt.xlabel('Year')
    plt.ylabel('Number')
    plt.title(crime)
    plt.xticks(rotation=90)  # Ruota le etichette sull'asse x di 90 gradi
    plt.grid(True)  # Aggiungi una griglia
    plt.show()
```
