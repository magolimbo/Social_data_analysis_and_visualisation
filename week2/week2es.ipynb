```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

data1 = pd.read_csv("data1.tsv", sep="\t", header=None, names=["x", "y"])
data2 = pd.read_csv("data2.tsv", sep="\t", header=None, names=["x", "y"])
data3 = pd.read_csv("data3.tsv", sep="\t", header=None, names=["x", "y"])
data4 = pd.read_csv("data4.tsv", sep="\t", header=None, names=["x", "y"])

print("Head of data1:")
print(data1.head())

print("Head of data2:")
print(data2.head())

print("Head of data3:")
print(data3.head())

print("Head of data4:")
print(data4.head())
```

## Access rows and columns of panda data frame and format
First assign a name to the columns of the data if they already don't have one, check above.

`print("{:10.4f}".format(x))`
- The empty string before the colon means "take the next provided argument to format()" – in this case the x as the only argument.
- The `10.4f` part after the colon is the format specification.
- The `f` denotes fixed-point notation.
- The `10` is the total width of the field being printed, left-padded by spaces.
- The `4` is the number of digits after the decimal point.


```python
data1_meany = data1["y"].mean()
data1_meanx = data1["x"].mean()
data1_var_y = data1["y"].var()
data1_var_x = data1["x"].var()

data2_meany = data2["y"].mean()
data2_meanx = data2["x"].mean()
data2_var_y = data2["y"].var()
data2_var_x = data2["x"].var()

data3_meany = data3["y"].mean()
data3_meanx = data3["x"].mean()
data3_var_y = data3["y"].var()
data3_var_x = data3["x"].var()

data4_meany = data4["y"].mean()
data4_meanx = data4["x"].mean()
data4_var_y = data4["y"].var()
data4_var_x = data4["x"].var()

for i in range(1, 5):
    print(f"*** Data {i} ***")
    print(f"y mean: {data1_meany:.2f}")
    print(f"x mean: {data1_meanx:.2f}")
    print(f"y variance: {data1_var_y:.2f}")
    print(f"x variance: {data1_var_x:.2f}")

```

## Pearson Correlation
 the Pearson correlation coefficient (PCC)[a] is a correlation coefficient that measures linear correlation between two sets of data. It is the ratio between the covariance of two variables and the product of their standard deviations. thus, it is essentially a normalized measurement of the covariance, such that the result always has a value between −1 and 1.

```python
cov_xy = []
pearson_correlation = []

cov_xy.append(np.sum((data1["y"] - data1_meany) * (data1["x"] - data1_meanx)) / (data1["y"].count()) )
pearson_correlation.append( cov_xy[0] / (math.sqrt(data1_var_y) * math.sqrt(data1_var_x)) )

cov_xy.append(np.sum((data2["y"] - data2_meany) * (data2["x"] - data2_meanx)) / (data2["y"].count()))
pearson_correlation.append( (cov_xy[1]) / (math.sqrt(data2_var_y) * math.sqrt(data2_var_x)))

cov_xy.append(np.sum((data3["y"] - data3_meany) * (data3["x"] - data3_meanx)) / (data3["y"].count()))
pearson_correlation.append(cov_xy[2] / (math.sqrt(data3_var_y) * math.sqrt(data3_var_x)))

cov_xy.append(np.sum((data4["y"] - data4_meany) * (data4["x"] - data4_meanx)) / (data4["y"].count()))
pearson_correlation.append(cov_xy[3] / (math.sqrt(data4_var_y) * math.sqrt(data4_var_x)))


for correlation in pearson_correlation:
    print(f"Pearson correlation: {correlation:.5f}")

```

## Linear Rergression

```python
from scipy import stats
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)
# Plot for data1
a, b, r_value, p_value, std_err = stats.linregress(data1["x"], data1["y"])
axs[0, 0].plot(data1["x"], a*data1["x"] + b, color="red")
axs[0, 0].scatter(data1["x"], data1["y"])
axs[0, 0].set_title("data 1")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("y")

# Plot for data2
a2, b2, r_value2, p_value2, std_err2 = stats.linregress(data2["x"], data2["y"])
axs[0, 1].plot(data2["x"], a2*data2["x"] + b2, color="red")
axs[0, 1].scatter(data2["x"], data2["y"])
axs[0, 1].set_title("data 2")
axs[0, 1].set_xlabel("x")
axs[0, 1].set_ylabel("y")

# Plot for data3
a3, b3, r_value3, p_value3, std_err3 = stats.linregress(data3["x"], data3["y"])
axs[1, 0].plot(data3["x"], a3*data3["x"] + b3, color="red")
axs[1, 0].scatter(data3["x"], data3["y"])
axs[1, 0].set_title("data 3")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("y")

# Plot for data4
a4, b4, r_value4, p_value4, std_err4 = stats.linregress(data4["x"], data4["y"])
axs[1, 1].plot(data4["x"], a4*data4["x"] + b4, color="red")
axs[1, 1].scatter(data4["x"], data4["y"])
axs[1, 1].set_title("data 4")
axs[1, 1].set_xlabel("x")
axs[1, 1].set_ylabel("y")

plt.tight_layout()
plt.show()
```

## distribution of crimes during the days, the months, the hours

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])

fig, axs = plt.subplots(2, 7, figsize=(40, 10))
#count occurrences during week
for i, focuscrime in enumerate(focuscrimes):
    spec_crimes = crimes[crimes['Category'] == focuscrime]
    crimesperday = spec_crimes['DayOfWeek'].value_counts()
    crimesperday = crimesperday.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']) 

    row = i // 7
    col = i % 7
    axs[row,col].bar(crimesperday.index, crimesperday.values)
    axs[row,col].set_title(focuscrime, fontsize = 20)
    axs[row,col].set_xlabel('Day of the week', fontsize = 15)
    axs[row,col].set_ylabel('Number of incidents', fontsize = 15)

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 7, figsize=(40, 10))
#count occurrences during week
for i, focuscrime in enumerate(focuscrimes):
    spec_crimes = crimes[crimes['Category'] == focuscrime]
    crimespermonth = pd.to_datetime(spec_crimes['Date']).dt.month.value_counts()
    crimespermonth = crimespermonth.reindex(range(1, 13)) 

    col = i if i < 7 else i-7
    row = 0 if i < 7 else 1
    axs[row,col].bar(crimespermonth.index, crimespermonth.values)
    axs[row,col].set_title(focuscrime, fontsize = 20)

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2,7, figsize=(40, 10))
for i, focuscrime in enumerate(focuscrimes):
    spec_crimes = crimes[crimes['Category'] == focuscrime]
    crimesperhour = spec_crimes['Time'].value_counts()
    crimesperhour = crimesperhour.reindex(range(0, 24)) 

    col = i if i < 7 else i-7
    row = 0 if i < 7 else 1
    axs[row,col].bar(crimesperhour.index, crimesperhour.values)
    axs[row,col].set_title(focuscrime, fontsize = 20)
```

## rules for good plots:
- use subplots
- space the subplots 
- label always x an y axis
- use meaningful counts on the axis (if you have to put the date on the x axis don't put the number of days elapsed from the beginning of the year)
- use the correct zoom level to avoid showing differences that are irrelevant
- ink-ratio the lowest possible
